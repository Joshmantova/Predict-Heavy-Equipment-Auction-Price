```python
df_preq = data[['ProductGroupDesc','SalePrice']].head(10)
# df_preq.head(10)

fig, ax = (plt.subplots(figsize=(18,12)))
barheights = df_preq['SalePrice'].values
barlabels = df_preq['ProductGroupDesc'].values
ax.bar(range(len(barheights)), barheights, color = 'green')
ax.set_xticks(range(len(barheights)))
ax.set_xticklabels(barlabels, rotation=90, size = 20)
ax.set_title('Top 20 Sale Prices for Equipment by Product Group', size = 40)
# ax.set_xlabel('Product Group', size = 20)
ax.set_ylabel('Sale Price in dollars', size= 20)
plt.tight_layout()
plt.savefig('salepricegraph.png')
```
